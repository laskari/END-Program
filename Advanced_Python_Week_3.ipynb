1. Write a function using only list filter lambda that can tell whether a number is a Fibonacci number or not. You can use a pre-calculated list/dict to store fab numbers till 10000 PTS:100


```
def fibonaci(n):
  fib=[0,1]
  for i in range(2, n):
    n3 = fib[-1] + fib[-2]
    if(n3>n):
      break
    else:
      fib.append(n3)
  return fib
fib = fibonaci(10000)

def verify_fab(fib, check):
  return list(filter(lambda x: x == check, fib))

result = verify_fab(fib,5)
print(bool(result))
```

a. add 2 iterables a and b such that a is even and b is odd


```
a = [x for x in range(50) if x%2 ==0]
b = [x for x in range(50) if x%2 ==1]
sum_list =[]
for x,y in zip(a,b):
   sum_list.append(x+y)

print(sum_list)
```

4. Using reduce functions: PTS:100

       add only even numbers in a list
       find the biggest character in a string (printable ascii characters)
       adds every 3rd number in a list

a. add only even numbers in a list


```
import functools
functools.reduce(lambda x,y: x+y, list(filter(lambda x: x%2==0, list(range(20)))))
```

b. find the biggest character in a string (printable ascii characters)

```
import functools
result = functools.reduce(lambda x, y: x if x > y else y, [x for x in 'welcome to school of ai'])
print(result, " = " ,ord(result))
```

c. adds every 3rd number in a list

```
array = list(range(1,40))
functools.reduce(lambda x, y: x+y, [x for x in array[2::3] ])
```

```
import random
def generate_random1():
    D1= random.randint(10,99)
    A1 = chr(random.randint(65, 90))
    A2 = chr(random.randint(65, 90))
    D2= random.randint(1000,10000)
    num_plate = 'KA'+str(D1)+A1+A2+str(D2)
    return num_plate

[generate_random1() for x in range(15) ]
```
